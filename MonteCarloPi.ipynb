```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Below is the randomness test for the random.uniform(), taking 10 million samples and testing the distribution of the points. The standard deviation is then taken. The average is within 1 percent of 50, being the expected average, as well as all error points being containted within two error bars provides strong evidence of there being an even distribution among all points

```python
def avg(data):
  x=0
  for i in data:
    x+=i
  return x/len(data)
N=10e6
i=0 #iterator for number generator
data=[]
dx=1/100
while i < N:
  num=random.uniform(0,1)
  value=np.floor(num/dx)
  data.append(value)
  i+=1

xbar= avg(data)
stdev=0
for a in data:
  stdev+=(a-xbar)**2/(len(data)-1)
stdev= np.sqrt(stdev)
print(f"average {xbar} with a standard deviation of {stdev}")
plt.hist(data,bins=100)
plt.xlabel('data bin')
plt.ylabel('data points per bin')
plt.title("Randomness Test for random library")
plt.show()
```
